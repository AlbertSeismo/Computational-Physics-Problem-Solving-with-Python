```python
#Page 144 Implementation: Bisection Algorithm
import matplotlib.pyplot as plt
from numpy import *
from scipy import interpolate 
%matplotlib notebook

#TASK 1&2: Visualizing the function
#The function to call tangent
def ftan(x):
    return sqrt(10-x)*tan(sqrt(10-x))-sqrt(x)

#Lets write the bisection method function for tangent
def bisection_tan(xplus, xminus,n,eps):
    for i in range (0,n):
        x = (xplus+xminus)/2
        print(i,x,ftan(x))     #This is where you had a mistake of f(x) instead of fb(x) so it was making your code run wrongly
        if (ftan(x)*ftan(xplus))>0 :
            xplus = x
        else :
            xminus = x
        if abs(ftan(x)) <= eps :
            print("The root is found with a value: ",x," with a precision = ",eps,ftan(x))
            return x
        if i == (n-1):
            print("The root is not found")
a = 6
b = 9.9
eps = 1e-6
n = 100
E = arange(a,b,0.05)

#Computing the root if it available (but remember tan has some singularities around the y-axis as x=0, tan(x)=sin(x)/cos(x))
#it approaches infinity
root = bisection_tan(a,b,n,eps)

#Lets visualize the function by plotting
plt.plot(E,ftan(E))
#plt.plot(x,f(x))
plt.grid()
plt.title("A plot of Tangent function with some singularities\n")
plt.xlabel("X")
plt.ylabel("ftan(x)")
plt.show()

"""In mathematics, a singularity is in general a point at which a given mathematical object is not defined, 
or a point where the mathematical object ceases to be well-behaved in some particular way, 
such as the lack of differentiability or analyticity. For example, the real function f(x)=1/x 
has a singularity at x=0, where it seems to "explode" to plus/minus infinity and is hence not defined """
```

```python
#Task 1 
x= array([0,25,50,75,100,125,150,175,200],float)     #E, Energy
y= array([10.6,16.0,45,83.5,52.8,19.9,10.8,8.25,4.7],float) #g, Cross-section scattering

l = len(x)   #length of the data
n = l-1 #polynomial degree
xp =float(input("Enter the value of E whose P is needed="))
P =0     #P capital initialize
for i in range(n+1):
    p=1   #initialize the product
    for j in range(n+1):
        if j !=i:
            p *= (xp-x[j])/(x[i]-x[j])
    P +=y[i]*p

print(P)   
```

```python
#Data given in the task for fitting using lagrange, from Table 7.1
E = array([0,25,50,75,100,125,150,175,200],float)
g = array([10.6,16.0,45,83.5,52.8,19.9,10.8,8.25,4.7],float)


#For the Lagrange method
E_Lagrange = linspace(E[0],E[-1],100)   #lets create a new dataset
g_Lagrange =array([],float)     #empty array

#lets apply a spline function on our function/plot   
#check scipy, spline
E_smooth = linspace(E.min(),E.max(),100)
g_interp = interpolate.splrep(E, g, s=0)
g_smooth = interpolate.splev(E_smooth, g_interp, der=0)

#see the slide above for the formula
for e in E_Lagrange:
    yp =0
    for xi,yi in zip(E,g):
        yp += yi*prod((e-E[E!=xi])/(xi-E[E!=xi]))    #applying conditional indices /index
    g_Lagrange = append(g_Lagrange,yp)
    

plt.figure(figsize=(10,5))    

plt.subplot(151)
plt.plot(E,g,"r-o")
plt.grid()
plt.title("Original data")
plt.xlabel("Energy, E /MeV")
plt.ylabel("Scattering Cross-section")

plt.subplot(153)
plt.plot(E_smooth,g_smooth, "-o")
plt.grid()
plt.title("Spline smoothed")
plt.xlabel("Energy, E /MeV")
plt.ylabel("Scattering Cross-section")

plt.subplot(155)
plt.plot(E_Lagrange,g_Lagrange,"-.")    #Lagrange fitting plot
plt.grid()
plt.xlabel("Energy, E /MeV")
plt.title("Lagrange fitting")
plt.ylabel("Scattering Cross-section")

plt.show()


```
